```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('diamonds.csv')
df.head()
```

```python
sns.set_style('whitegrid') 
```

```python
df = df.drop('Unnamed: 0', axis=1)
df.head()
```

```python
dfCar0 = df[df['carat']<1]
dfCar0.head()
```

```python
dfCar1 = df[(df['carat']<2)&(df['carat']>=1)]
dfCar1.head()
```

```python
dfCar2 = df[(df['carat']<3)&(df['carat']>=2)]
dfCar2.head()
```

```python
dfCar3 = df[(df['carat']<4)&(df['carat']>=3)]
dfCar3.head()
```

```python
dfCar4 = df[(df['carat']<5)&(df['carat']>=4)]
dfCar4.head()
```

```python
dfCar5 = df[(df['carat']<6)&(df['carat']>=5)]
dfCar5.head()
```

```python
print('Banyaknya data ber carat 0 : ', len(dfCar0))
print('Banyaknya data ber carat 1 : ', len(dfCar1))
print('Banyaknya data ber carat 2 : ', len(dfCar2))
print('Banyaknya data ber carat 3 : ', len(dfCar3))
print('Banyaknya data ber carat 4 : ', len(dfCar4))
print('Banyaknya data ber carat 5 : ', len(dfCar5))
```

```python
print('Mean dari price ber carat 0 : ', dfCar0['price'].mean())
print('Mean dari price ber carat 1 : ', dfCar1['price'].mean())
print('Mean dari price ber carat 2 : ', dfCar2['price'].mean())
print('Mean dari price ber carat 3 : ', dfCar3['price'].mean())
print('Mean dari price ber carat 4 : ', dfCar4['price'].mean())
print('Mean dari price ber carat 5 : ', dfCar5['price'].mean())
```

```python
carat0 = []
carat1 = []
carat2 = []
carat3 = []
carat4 = []
carat5 = []

for i in range(1,34881):
    carat0.append('Zero')
for i in range(1,16907):
    carat1.append('One')
for i in range(1,2115):
    carat2.append('Two')
for i in range(1,35):
    carat3.append('Three')
for i in range(1,6):
    carat4.append('Four')
for i in range(1,2):
    carat5.append('Five')
```

```python
dfCar0['carat_category'] = carat0
dfCar0.head()
```

```python
dfCar1['carat_category'] = carat1
dfCar1.head()
```

```python
dfCar2['carat_category'] = carat2
dfCar2.head()
```

```python
dfCar3['carat_category'] = carat3
dfCar3.head()
```

```python
dfCar4['carat_category'] = carat4
dfCar4.head()
```

```python
dfCar5['carat_category'] = carat5
dfCar5.head()
```

```python
frames = [dfCar0, dfCar1, dfCar2, dfCar3, dfCar4, dfCar5]
```

```python
diamonds = pd.concat(frames)
diamonds = diamonds.sort_index()
```

```python
diamonds.head()
```

```python
df = diamonds[['carat', 'carat_category', 'cut', 'color', 'clarity', 'depth', 'table', 'x', 'y', 'z', 'price']]
df.head()
```

```python
len(df)
```

```python
sns.factorplot(x='carat_category',y='price',data=df, kind='box')
```

# Cleaning Outliers

```python
dfOutliersCarat = pd.DataFrame({'id': i,
                           'mean_price': df[df['carat_category']==i]['price'].mean(),
                           'std_price': df[df['carat_category']==i]['price'].std(),
                           'batas_min_price': df[df['carat_category']==i]['price'].mean()-(df[df['carat_category']==i]['price'].std())*3,
                           'batas_max_price': df[df['carat_category']==i]['price'].mean()+(df[df['carat_category']==i]['price'].std())*3,
                          }for i in df['carat_category'].unique())
dfOutliersCarat
```
