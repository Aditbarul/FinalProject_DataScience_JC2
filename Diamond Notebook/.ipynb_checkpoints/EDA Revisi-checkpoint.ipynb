```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.plotly as py
import plotly.graph_objs as go

sns.set_style('whitegrid')    
%matplotlib inline
```

```python
df = pd.read_csv('diamonds.csv')
df.head()
```

```python
dfCar0 = df[df['carat']<1]
dfCar1 = df[(df['carat']<2)&(df['carat']>=1)]
dfCar2 = df[(df['carat']<3)&(df['carat']>=2)]
dfCar3 = df[(df['carat']<4)&(df['carat']>=3)]
dfCar4 = df[(df['carat']<5)&(df['carat']>=4)]
dfCar5 = df[(df['carat']<6)&(df['carat']>=5)]

carat0 = []
carat1 = []
carat2 = []
carat3 = []
carat4 = []
carat5 = []

for i in range(1,34881):
    carat0.append('Zero')
for i in range(1,16907):
    carat1.append('One')
for i in range(1,2115):
    carat2.append('Two')
for i in range(1,35):
    carat3.append('Three')
for i in range(1,6):
    carat4.append('Four')
for i in range(1,2):
    carat5.append('Five')
    
dfCar0['carat_category'] = carat0
dfCar1['carat_category'] = carat1
dfCar2['carat_category'] = carat2
dfCar3['carat_category'] = carat3
dfCar4['carat_category'] = carat4
dfCar5['carat_category'] = carat5
```

```python
frames = [dfCar0, dfCar1, dfCar2, dfCar3, dfCar4, dfCar5]
diamonds = pd.concat(frames)
diamonds = diamonds.sort_index()
df = diamonds[['carat', 'carat_category', 'cut', 'color', 'clarity', 'depth', 'table', 'x', 'y', 'z', 'price']]
df.head()
```

```python
len(df[df['cut']=='Fair'])
```

```python
len(df[df['cut']=='Ideal'])
```

```python
df['cut'] = pd.Categorical(values=df.cut, categories=['Fair', 'Good', 'Very Good', 'Premium', 'Ideal'], ordered=True)
df['color'] = pd.Categorical(values = df.color, categories=['D', 'E', 'F', 'G', 'H','I','J'], ordered=True)
df['clarity'] = pd.Categorical(values = df.clarity, categories=['I1', 'SI2', 'SI1','VS2','VS1','VVS2','VVS1','IF'], ordered=True)

plt.figure(figsize = (12,8), dpi=400)
plt.subplot(3,1,1)
sns.countplot(y='cut', data=df, color="c")
plt.xlim((0,22000))
plt.xlabel('')
plt.ylabel('Cut')

plt.subplot(3,1,2)
sns.countplot(y='color', data=df, color="r")
plt.ylabel('Color')
plt.xlim((0,22000))
plt.xlabel('')

plt.subplot(3,1,3)
sns.countplot(y='clarity', data=df)
plt.xlabel('Number of diamonds in the dataset')
plt.ylabel('Clarity')
plt.xlim((0,22000))

plt.tight_layout()
plt.show()
```

```python
color_range = ['D', 'E', 'F', 'G', 'H','I','J']
plt.style.use('fivethirtyeight')
plt.figure(figsize = (12,8), dpi=200)
for col in color_range:   
    plt.scatter(df[df['color']== col].carat, df[df['color']== col].price, alpha=0.8, marker = 'o', label=col)   
plt.xlabel('Carat')
plt.ylabel('Price, $')
plt.legend()
plt.title('''Distribution of diamond's price via carat and color''')
plt.savefig('''Distribution of diamond's price via carat and color''')
plt.show()
```

```python
clarity_range = ['SI2', 'SI1', 'VS1', 'VS2', 'VVS2', 'VVS1', 'I1', 'IF']
plt.style.use('fivethirtyeight')
plt.figure(figsize = (12,8), dpi=200)
for cal in clarity_range:   
    plt.scatter(df[df['clarity']== cal].carat , df[df['clarity']== cal].price, alpha=0.8, marker = 'D', label=cal)   
plt.xlabel('Carat')
plt.ylabel('Price, $')
plt.legend()
plt.title('''Distribution of diamond's price via carat and clarity''')
plt.savefig('''Distribution of diamond's price via carat and clarity''')
plt.show()
```
