```python
# Mencari Outliers dari Price menggunakan 3 std
print('Mean dari Price', df['price'].mean())
print('\n')
print('Mean dari Price', df['price'].std())
print('\n')
print('Batas Min Price', df['price'].mean()-(df['price'].std())*3)
print('\n')
print('Batas Max Price', df['price'].mean()+(df['price'].std())*3)

max_3std = df['price'].mean()+(df['price'].std())*3
```

```python
# Mencari Outliers dari Price menggunakan 2 std
print('Mean dari Price', df['price'].mean())
print('\n')
print('Mean dari Price', df['price'].std())
print('\n')
print('Batas Min Price', df['price'].mean()-(df['price'].std())*2)
print('\n')
print('Batas Max Price', df['price'].mean()+(df['price'].std())*2)

max_2std = df['price'].mean()+(df['price'].std())*2
```

```python
# Mencari Outliers dari Price menggunakan 1 std
print('Mean dari Price', df['price'].mean())
print('\n')
print('Mean dari Price', df['price'].std())
print('\n')
print('Batas Min Price', df['price'].mean()-(df['price'].std())*1)
print('\n')
print('Batas Max Price', df['price'].mean()+(df['price'].std())*1)

max_1std = df['price'].mean()+(df['price'].std())*1
```

```python
dfOutliersCut = pd.DataFrame({'id': i,
                           'mean_fare': df[df['cut']==i]['price'].mean(),
                           'std_fare': df[df['cut']==i]['price'].std(),
                           'batas_min_fare': df[df['cut']==i]['price'].mean()-(df[df['cut']==i]['price'].std())*3,
                           'batas_max_fare': df[df['cut']==i]['price'].mean()+(df[df['cut']==i]['price'].std())*3,
                          }for i in df['cut'].unique())
dfOutliersCut
```

```python
dfOutliersColor = pd.DataFrame({'id': i,
                           'mean_fare': df[df['color']==i]['price'].mean(),
                           'std_fare': df[df['color']==i]['price'].std(),
                           'batas_min_fare': df[df['color']==i]['price'].mean()-(df[df['color']==i]['price'].std())*3,
                           'batas_max_fare': df[df['color']==i]['price'].mean()+(df[df['color']==i]['price'].std())*3,
                          }for i in df['color'].unique())
dfOutliersColor
```

```python
dfOutliersClarity = pd.DataFrame({'id': i,
                           'mean_fare': df[df['clarity']==i]['price'].mean(),
                           'std_fare': df[df['clarity']==i]['price'].std(),
                           'batas_min_fare': df[df['clarity']==i]['price'].mean()-(df[df['clarity']==i]['price'].std())*3,
                           'batas_max_fare': df[df['clarity']==i]['price'].mean()+(df[df['clarity']==i]['price'].std())*3,
                          }for i in df['clarity'].unique())
dfOutliersClarity
```
